```python
import pandas as pd
import matplotlib.pyplot as plt

static_path = "Results\static_tripinfo.csv"

df = pd.read_csv(static_path)
size = 900  #bin size
df.head()
```

```python
bins = pd.cut(df['tripinfo_arrival'], np.arange(0,df['tripinfo_arrival'].iloc[-1], size), labels = (np.arange(0,df['tripinfo_arrival'].iloc[-1], size)/60)[1:])
bins = bins.cat.add_categories(max(bins) + size/60).fillna(max(bins) + size/60)
df['bins'] = bins

df_quarter_mean = df.groupby('bins').mean()
df_quarter_mean
```

## Waiting Time

```python
fig, ax = plt.subplots(figsize=[15,8])


x = df_quarter_mean.index
y = df_quarter_mean['tripinfo_waitingTime']

ax.plot(x,y)
ax.set_xlabel("Simulation Time in Minutes (m)")
ax.set_ylabel("Waiting Time in Seconds (s)")
ax.set_title("Waiting Time for Vehicles Using Static Traffic Light System")

plt.show()
```

## Time Loss

```python
fig, ax = plt.subplots(figsize=[15,8])


x = df_quarter_mean.index
y = df_quarter_mean['tripinfo_timeLoss']

ax.plot(x,y)
ax.set_xlabel("Simulation Time in Minutes (m)")
ax.set_ylabel("Time Loss in Seconds (s)")
ax.set_title("Time Loss for Vehicles Using Static Traffic Light System")

plt.show()
```
