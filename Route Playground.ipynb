```python
import pandas as pd

df = pd.read_excel("sample_data.xlsx", index_col=0)
df
```

```python
from_val = []
to_val = []
count = []
for idx_r, row in enumerate(df.index.to_list()):
    for idx_c, column in enumerate(df.columns.to_list()):
        print(row, column)
        from_val.append(df.index.to_list()[idx_r])
        to_val.append(df.columns.to_list()[idx_c])
        count.append(df.iloc[idx_r, idx_c])

new_df = pd.DataFrame({'From': from_val,'To': to_val, 'Count': count})
new_df
```

```python
from_val, to_val, count_val = new_df.iloc[0,:].to_list()

from IPython.display import display

display(from_val, to_val, count_val)
```
